```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
lambda1 = np.linspace(0, 1, 50)
lambda2 = np.array([1-i for i in lambda1])
```

```python
x1 = 0
x2 = 1
y1 = 0
y2 = 1
```

```python
plt.scatter(lambda1*x1 + lambda2*x2, lambda1*y1 + lambda2*y2)
plt.show()
```
