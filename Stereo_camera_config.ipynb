```python
import numpy as np
import matplotlib.pyplot as plt
```

Design of a 4 camera system for 60cm range under 20 cm depth error.

```python
# z = b*f/d = m*pixels/pixels
# v = dx/dt = m/s
```

```python
# Stereo error

ed = 0.75 # error in disparity
baseline = 0.1
focal_length = 150
# Ax^2 + Bx + C
# (ed/(baseline*focal_length))z^2
# A = ed/(baseline*focal_length)
# B = 0, C = 0
zrange = np.arange(0.01, 30.00, 0.5) # range of disparity

ez1 = [(z*z*ed)/(baseline*focal_length) for z in zrange]

fig, axes = plt.subplots(3,1)
plt.axis('equal')
axes[0].plot(zrange, ez1)
axes[0].set(ylabel='Error (m)')
axes[0].set_title(f'Baseline : {baseline} m')


baseline = 0.2
ez2 = [(z*z*ed)/(baseline*focal_length) for z in zrange]
axes[1].plot(zrange, ez2)
axes[1].set(ylabel='Error (m)')
axes[1].set_title(f'Baseline : {baseline} m')

baseline = 0.3
ez3 = [(z*z*ed)/(baseline*focal_length) for z in zrange]
axes[2].plot(zrange, ez3)
axes[2].set(xlabel='Distance (m)')
axes[2].set(ylabel='Error (m)')
axes[2].set_title(f'Baseline : {baseline} m')
```

To design a 3 camera system for 5 m range under 3 cm depth error.

```python
# Stereo error

ed = 0.2 # error in disparity
baseline = 0.4
focal_length = 150

zrange = np.arange(0.01, 3.50, 0.5) # range of depth

ez1 = [(z*z*ed)/(baseline*focal_length) for z in zrange]

fig, axes = plt.subplots(2,1)

axes[0].plot(zrange, ez1)
axes[0].set(ylabel='Error (m)')
axes[0].set_title(f'Baseline : {baseline} m')

baseline = 3
ez2 = [(z*z*ed)/(baseline*focal_length) for z in zrange]
axes[1].plot(zrange, ez2)
axes[1].set(ylabel='Error (m)')
axes[1].set_title(f'Baseline : {baseline} m')

```
